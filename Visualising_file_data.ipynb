# Visualising file data as data frames

```python
import matplotlib.pyplot as plt

from media_tools import *
import pandas as pd
```

### Image file data:

```python
image = image_tools.read_image("sample_media/Mosse_2012_Hot_Rats.jpg")

blue_channel = image[:,:,0]
green_channel = image[:,:,1]
red_channel = image[:,:,2]
```

#### Visualising image channels:
We can show each channel of an image by calling the `image_tools` package's `show_image(data)` function for the image
channel data like so:

```python
image_tools.show_image(blue_channel, title="Blue channel")
image_tools.show_image(green_channel, title="Green channel")
image_tools.show_image(red_channel, title="Red channel")
```
