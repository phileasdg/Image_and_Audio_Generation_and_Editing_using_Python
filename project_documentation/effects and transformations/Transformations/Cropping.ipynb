# Cropping 2d arrays

```python
from media_tools import *
import cv2

sample_media = "C:/Users/pdazeley-gaist23/PycharmProjects/Image_and_Audio_Generation_and_Editing_using_Python/sample_media/"
image = image_tools.read_image(sample_media + "student_guide.jpg")
```

You can crop a 2d array using the `crop_image` function of the `image_tools` module from the `media_tools` package, which takes
as its arguments a 2d array, and two tupples, the coordinates of the top left, and bottom right of the desired crop area
of the source array.

```python
# show the start image
image_tools.show_image(image, axis=True, title="Original image", scale_ratio=1)

# crop an image to a range described by the top left and bottom right corners
image_cropped = image_tools.crop_image(image, top_left=(700, 400), bottom_right=(1400, 2400))
image_tools.show_image(image_cropped, axis=True, title="Cropped image", scale_ratio=1)
```
